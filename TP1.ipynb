## *API pour les mots :*

```python
import requests

def obtenir_mots_aleatoires(nombre_mots):
    random_words = []
    api_url = f'https://random-word-api.herokuapp.com/word?number={nombre_mots}'
    response = requests.get(api_url)
    mots = response.json()
    for mot in mots:
        random_words.append(mot)
    return random_words

def obtenir_mots_aleatoires_differents(nombres_mots2,mots):
    different_words = []
    mots2 = obtenir_mots_aleatoires(nombres_mots2)
    for mot in mots2:
        while mot in mots:
            mot = obtenir_mots_aleatoires(1)
        different_words.append(mot)
    return different_words
  

nombre_mots = 10000 
random_words = obtenir_mots_aleatoires(nombre_mots)


nombres_mots2 = 100
random_words2 = obtenir_mots_aleatoires_differents(nombres_mots2,random_words)


```

## *Fonction de hachage d'un mot :*

```python
def hachage_multiple_mot(mot,theta,taille_filtre):
    hach =[]*len(theta)
    byte_mot = bytes(mot, 'utf-8')
    int_byte = int.from_bytes(byte_mot, byteorder='big')
    for i in range(len(theta)):
        r = int_byte * theta[i]
        r = r - int(r)
        r = r*taille_filtre
        r = int(r)
        hach.append(r)
    return hach   
```

## *Filtre de Bloom :*

### *Définition paramètres :*

```python
theta = [0.24, 0.32, 0.16, 0.84, 0.77, 0.44]
taille_filtre = 900
```

### *Initialisation du filtre :*

```python
filtre_bloom = [0]*taille_filtre
```

### *Remplissage :*

```python
def hachage_multiple_liste_mots(random_words):
    for mot in random_words:
        hach = hachage_multiple_mot(mot,theta,taille_filtre)
        for i in hach:
            filtre_bloom[i] = 1
    return filtre_bloom

hachage_multiple_liste_mots(random_words)
```
